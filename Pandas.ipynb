## **PANDAS**

### **ANALYSING THE DATA**

To read and compute the data we must first convert the excel file to a csv file using ***to_csv()*** function

```python
import pandas as pd
import matplotlib.pyplot as plt


# Read the Excel file into a DataFrame
excel_file = '/Dry_Bean_Dataset.xlsx'
df = pd.read_excel(excel_file)

# Save the DataFrame to a CSV file
csv_file = 'output.csv'
df.to_csv(csv_file, index=False)

df.head()
```

To display basic information about the DataFrame

```python

print(df.info())

```

To display the summary statitistics of the DataFrame

```python

print(df.describe())
```

### **CLEANING THE DATA**

This can be achieved by dropping rows with missing values as shown in the code spinnet below

```python
new_df = df.dropna()

new_df.head()
```

### **Replace Empty Values**

Another way of dealing with empty cells is to insert a new value instead.

This way you do not have to delete entire rows just because of some empty cells.

The ***fillna()*** method allows us to replace empty cells with a value:

```python
df.fillna(130, inplace = True)
```

### **PLOTTING THE DATA**

1. **HISTOGRAM**


A ***histogram*** is a graphical representation of the distribution of numerical data. It consists of bars that show the frequency or count of data points falling into different intervals or bins. Histograms are useful for visualizing the shape, center, and spread of a dataset, as well as identifying patterns and outliers.

```python
# Histogram of the 'Area' column
df['Area'].plot(kind='hist', bins=20, edgecolor='black')
plt.xlabel('Area')
plt.ylabel('Frequency')
plt.title('Histogram of Area')
plt.show()

```

2. **SCATTER PLOT**


A ***scatter plot*** is a type of plot that displays the relationship between two continuous variables. Each data point in the plot represents a single observation, with one variable plotted on the x-axis and the other variable plotted on the y-axis. Scatter plots are helpful for visualizing correlations, trends, and patterns between variables, and they can indicate the strength and direction of the relationship between the variables.

```python
# Scatter plot of 'Area' vs 'Perimeter'
plt.scatter(df['Area'], df['Perimeter'], alpha=0.5)
plt.xlabel('Area')
plt.ylabel('Perimeter')
plt.title('Scatter Plot of Area vs Perimeter')
plt.show()
```
